```python
import requests
from PIL import Image
import cv2
import numpy as np
from io import BytesIO
import json
```

```python
CONVERTER_API_ENDPOINT = AWS_LAMBDA_ENPOINT
```

```python
# the convert function takes an img file and API endpoint as inputs, when you call the function the image is going to AWS
# LAMBDA function for image processing, the processed image will converted to np array at the end of the function and returned to the client.
```

```python
def convert(img, converter_api):
    API_ENDPOINT = converter_api

    is_success, im_buf_arr = cv2.imencode(".png", img)
    byte_im = im_buf_arr.tobytes()

    r = requests.post(url=API_ENDPOINT, data=byte_im)

    img_ = Image.open(BytesIO(r.content))

    return np.asarray(img_)
```
